## Q5. Date Transformation
Extract the year from the date_added column.

```python
df["date_added"] = df["date_added"].str.strip()

df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")

df["year_added"] = df["date_added"].dt.year

df["year_added"].value_counts().head()
```

Markdown Analysis
The date_added column was converted to datetime format using pd.to_datetime().
The year was then extracted using the dt.year attribute.

This allows us to analyze how many titles were added to Netflix each year.

## Q6. Boolean Indexing
Filter shows with TV-MA rating that were added after 2018.

```python
filtered_shows = df[(df["rating"]=="TV-MA") & (df["year_added"]>2018)]

filtered_shows.head()
```

Markdown Analysis


Boolean indexing was used to filter rows where the rating is TV-MA
and the year_added is greater than 2018.

This helps identify mature-rated content that has been added in
recent years.

## Q7. Duration Classification using apply()
Categorize content based on duration length.

```python
def classify_duration(duration):
    if "min" in str(duration):
        minutes = int(duration.split()[0])
        if minutes < 30:
            return "Short"
        elif minutes <= 90:
            return "Standard"
        else:
            return "Long"
    else:
        return "Series"

df["duration_category"] = df["duration"].apply(classify_duration)

df["duration_category"].value_counts()
```

Markdown Analysis
A custom function was created to categorize content duration.
The apply() function applied this classification to every row
in the duration column.

This helps identify how many titles fall into Short, Standard,
Long, or Series categories.

## Q8. Index Alignment
Combine multiple Series objects.

```python
country_count = df["country"].value_counts()

duration_count = df.groupby("country")["duration"].count()

combined_df = pd.concat([country_count, duration_count], axis=1)

combined_df.head()
```

Markdown Analysis
Two Series were created: one showing the number of titles per country
and another counting the durations.

Using pd.concat(), both Series were merged based on their index.
Pandas automatically aligns the data using the country names.
